# Context Reveals Signal

This notebook synthesizes the repository.

Previous notebooks showed:

- shared noise can cancel

The central lesson is:

**Context is shared structure that allows apparent noise to be separated from signal.**


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

ROOT = Path("..")
FIGURES = ROOT / "figures"
DATA = ROOT / "data"

FIGURES.mkdir(exist_ok=True)
DATA.mkdir(exist_ok=True)
```

## Context and error

As context increases, uncertainty decreases.

This is not a specific physical law.

It is a conceptual model illustrating how shared structure improves recovery.


```
context = np.linspace(0, 1, 200)

error = 1 / (1 + 8 * context)

plt.figure(figsize=(7,4))

plt.plot(context, error, linewidth=3)

plt.xlabel("Context strength")
plt.ylabel("Relative error")

plt.title("More context enables lower error")

plt.tight_layout()

plt.savefig(
    FIGURES / "43_context_error_curve.png",
    dpi=200
)

plt.show()
```

## Context map

Different domains use different forms of context.

The common theme is that shared structure reduces apparent randomness.


```
domains = pd.DataFrame({
    "domain": [
        "atom interferometry",
        "sensor arrays",
        "forecasting",
        "continual learning"
    ],
    "apparent_problem": [
        "laser phase noise",
        "spatial disturbance",
        "prediction error",
        "forgetting"
    ],
    "context": [
        "common-mode phase",
        "geometry",
        "latent state",
        "residue memory"
    ],
    "recovered_structure": [
        "differential signal",
        "measurable modes",
        "future state",
        "retained knowledge"
    ]
})

domains
```

```
domains.to_csv(
    DATA / "43_domains_table.csv",
    index=False
)
```

```
fig, ax = plt.subplots(figsize=(10,4))

ax.axis("off")

table = ax.table(
    cellText=domains.values,
    colLabels=domains.columns,
    loc="center"
)

table.auto_set_font_size(False)
table.set_fontsize(9)
table.scale(1, 1.5)

plt.title("Context reveals structure across domains")

plt.tight_layout()

plt.savefig(
    FIGURES / "43_context_map.png",
    dpi=200,
    bbox_inches="tight"
)

plt.show()
```

## Error reduction summary

The previous notebooks can be viewed as examples of context-enabled error reduction.


```
categories = [
    "shared noise",
    "correlation",
    "baseline",
    "geometry",
    "differential readout"
]

relative_error = [
    1.0,
    0.75,
    0.55,
    0.35,
    0.15
]

plt.figure(figsize=(8,4))

plt.plot(
    categories,
    relative_error,
    marker="o",
    linewidth=2
)

plt.ylabel("Relative error")

plt.title("Increasing structure improves recovery")

plt.tight_layout()

plt.savefig(
    FIGURES / "43_error_reduction_summary.png",
    dpi=200
)

plt.show()
```
